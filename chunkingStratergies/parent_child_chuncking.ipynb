```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from langchain_text_splitters import RecursiveCharacterTextSplitter
import uuid

document = open("document.txt").read()

parent_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1500,
    chunk_overlap=200
)

child_splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=50
)

parents = {}
children = []

for parent_text in parent_splitter.split_text(document):

    parent_id = str(uuid.uuid4())

    parents[parent_id] = parent_text

    for child_text in child_splitter.split_text(parent_text):

        children.append({
            "parent_id": parent_id,
            "child_text": child_text
        })

model = SentenceTransformer(
    "BAAI/bge-small-en-v1.5",
    device="cpu"
)

child_embeddings = model.encode(
    [c["child_text"] for c in children],
    normalize_embeddings=True
)

query = "What is the A18 chip?"

query_embedding = model.encode(
    query,
    normalize_embeddings=True
)

scores = cosine_similarity(
    [query_embedding],
    child_embeddings
)[0]

best_idx = scores.argmax()

best_child = children[best_idx]

parent_context = parents[
    best_child["parent_id"]
]

print(parent_context)
```
